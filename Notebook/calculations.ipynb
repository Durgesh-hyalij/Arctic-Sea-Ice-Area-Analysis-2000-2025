```python
import os
import numpy as np
import pandas as pd

from scipy.stats import pearsonr
import pymannkendall as mk
```

```python
PROJECT = r"C:\Users\ACER\OneDrive\Desktop\ISRO\snow_2"

MASTER_DATASET = os.path.join(
    PROJECT,
    "Master_dataset"
)

RESULTS = os.path.join(
    PROJECT,
    "Results"
)
```

```python
SEAICE_FILE = os.path.join(
    MASTER_DATASET,
    "Yearly_Average_Sea_Ice.xlsx"
)

ALBEDO_FILE = os.path.join(
    MASTER_DATASET,
    "Albedo_Statistics_2000_2026 (1).xlsx"
)

seaice = pd.read_excel(SEAICE_FILE)

albedo = pd.read_excel(
    ALBEDO_FILE,
    sheet_name="Yearly_albedo_statistics"
)

print(seaice.head())

print(albedo.head())
```

```python
# ============================================================
# CREATE MASTER DATASET
# ============================================================

df = pd.merge(
    seaice,
    albedo,
    on="Year",
    how="inner"
)

print(df.head())

print()

print(df.shape)

print()

print(df.columns.tolist())
```

```python
# ============================================================
# LONG-TERM MEANS
# ============================================================

sea_long_mean = df["Yearly Mean"].mean()

albedo_long_mean = df["Ocean Mean Albedo"].mean()

print(f"Sea Ice Long-Term Mean      : {sea_long_mean:.4f}")

print(f"Ocean Albedo Long-Term Mean : {albedo_long_mean:.4f}")
```

```python
# ============================================================
# PERCENTAGE DEVIATIONS
# ============================================================

df["Sea Ice Deviation (%)"] = (
    (df["Yearly Mean"] - sea_long_mean)
    / sea_long_mean
) * 100

df["Albedo Deviation (%)"] = (
    (df["Ocean Mean Albedo"] - albedo_long_mean)
    / albedo_long_mean
) * 100
```

```python
print(df.head())

print()

print(df.shape)

print()

print(df.columns.tolist())
```

```python
# ============================================================
# DESCRIPTIVE STATISTICS
# ============================================================

descriptive_df = pd.DataFrame({

    "Statistic":[
        "Count",
        "Mean",
        "Median",
        "Standard Deviation",
        "Variance",
        "Minimum",
        "25th Percentile (Q1)",
        "75th Percentile (Q3)",
        "Interquartile Range (IQR)",
        "Maximum",
        "Range"
    ],

    "Sea Ice (Million km²)":[

        df["Yearly Mean"].count(),

        df["Yearly Mean"].mean(),

        df["Yearly Mean"].median(),

        df["Yearly Mean"].std(),

        df["Yearly Mean"].var(),

        df["Yearly Mean"].min(),

        df["Yearly Mean"].quantile(0.25),

        df["Yearly Mean"].quantile(0.75),

        df["Yearly Mean"].quantile(0.75) -
        df["Yearly Mean"].quantile(0.25),

        df["Yearly Mean"].max(),

        df["Yearly Mean"].max() -
        df["Yearly Mean"].min()

    ],

    "Ocean Mean Albedo":[

        df["Ocean Mean Albedo"].count(),

        df["Ocean Mean Albedo"].mean(),

        df["Ocean Mean Albedo"].median(),

        df["Ocean Mean Albedo"].std(),

        df["Ocean Mean Albedo"].var(),

        df["Ocean Mean Albedo"].min(),

        df["Ocean Mean Albedo"].quantile(0.25),

        df["Ocean Mean Albedo"].quantile(0.75),

        df["Ocean Mean Albedo"].quantile(0.75) -
        df["Ocean Mean Albedo"].quantile(0.25),

        df["Ocean Mean Albedo"].max(),

        df["Ocean Mean Albedo"].max() -
        df["Ocean Mean Albedo"].min()

    ]

})

descriptive_df
```

```python
# ============================================================
# PEARSON CORRELATION
# ============================================================

from scipy.stats import pearsonr

r, p = pearsonr(
    df["Sea Ice Deviation (%)"],
    df["Albedo Deviation (%)"]
)

print("="*50)
print("PEARSON CORRELATION")
print("="*50)

print(f"Correlation Coefficient (r) : {r:.6f}")
print(f"P-value                    : {p:.6f}")
```

```python
# ============================================================
# RELATIONSHIP STRENGTH
# ============================================================

if abs(r) >= 0.90:
    relationship = "Very Strong"
elif abs(r) >= 0.70:
    relationship = "Strong"
elif abs(r) >= 0.50:
    relationship = "Moderate"
elif abs(r) >= 0.30:
    relationship = "Weak"
else:
    relationship = "Very Weak"

direction = "Positive" if r > 0 else "Negative"

print(f"Relationship : {relationship} {direction}")
```

```python
# ============================================================
# PEARSON CORRELATION TABLE
# ============================================================

correlation_df = pd.DataFrame({

    "Variable 1":[
        "Sea Ice Deviation (%)"
    ],

    "Variable 2":[
        "Ocean Mean Albedo Deviation (%)"
    ],

    "Pearson Correlation (r)":[
        r
    ],

    "P-value":[
        p
    ],

    "Relationship":[
        f"{relationship} {direction}"
    ],

    "Interpretation":[
        "Statistically Significant" if p < 0.05
        else "Not Statistically Significant"
    ]

})

correlation_df
```

```python
# ============================================================
# MANN-KENDALL TEST
# ============================================================

import pymannkendall as mk

sea_result = mk.original_test(
    df["Sea Ice Deviation (%)"]
)

alb_result = mk.original_test(
    df["Albedo Deviation (%)"]
)
```

```python
# ============================================================
# MANN-KENDALL RESULTS TABLE
# ============================================================

mk_df = pd.DataFrame({

    "Variable":[
        "Sea Ice",
        "Ocean Mean Albedo"
    ],

    "Trend":[
        sea_result.trend,
        alb_result.trend
    ],

    "Significant (H)":[
        sea_result.h,
        alb_result.h
    ],

    "P-value":[
        sea_result.p,
        alb_result.p
    ],

    "Z Statistic":[
        sea_result.z,
        alb_result.z
    ],

    "Kendall Tau":[
        sea_result.Tau,
        alb_result.Tau
    ],

    "S Statistic":[
        sea_result.s,
        alb_result.s
    ],

    "Variance(S)":[
        sea_result.var_s,
        alb_result.var_s
    ],

    "Sen's Slope (%/year)":[
        sea_result.slope,
        alb_result.slope
    ],

    "Intercept":[
        sea_result.intercept,
        alb_result.intercept
    ]

})

mk_df
```

```python
# ============================================================
# TREND STATISTICS SUMMARY
# ============================================================

trend_df = pd.DataFrame({

    "Parameter":[

        "Long-Term Mean",

        "Mean Deviation (%)",

        "Maximum Deviation (%)",

        "Minimum Deviation (%)",

        "Sen's Slope (%/year)",

        "Kendall Tau",

        "P-value",

        "Trend",

        "Significant"

    ],

    "Sea Ice":[

        df["Yearly Mean"].mean(),

        df["Sea Ice Deviation (%)"].mean(),

        df["Sea Ice Deviation (%)"].max(),

        df["Sea Ice Deviation (%)"].min(),

        sea_result.slope,

        sea_result.Tau,

        sea_result.p,

        sea_result.trend,

        sea_result.h

    ],

    "Ocean Mean Albedo":[

        df["Ocean Mean Albedo"].mean(),

        df["Albedo Deviation (%)"].mean(),

        df["Albedo Deviation (%)"].max(),

        df["Albedo Deviation (%)"].min(),

        alb_result.slope,

        alb_result.Tau,

        alb_result.p,

        alb_result.trend,

        alb_result.h

    ]

})

trend_df
```

```python
# ============================================================
# EXPORT FINAL EXCEL WORKBOOK
# ============================================================

OUTPUT_FILE = os.path.join(
    RESULTS,
    "Arctic_SeaIce_Albedo_Trend_Analysis_2000_2025.xlsx"
)

with pd.ExcelWriter(
    OUTPUT_FILE,
    engine="openpyxl"
) as writer:

    # ========================================================
    # Sheet 1
    # ========================================================
    df.to_excel(
        writer,
        sheet_name="01_Master_Dataset",
        index=False
    )

    # ========================================================
    # Sheet 2
    # ========================================================
    descriptive_df.to_excel(
        writer,
        sheet_name="02_Descriptive_Statistics",
        index=False
    )

    # ========================================================
    # Sheet 3
    # ========================================================
    correlation_df.to_excel(
        writer,
        sheet_name="03_Pearson_Correlation",
        index=False
    )

    # ========================================================
    # Sheet 4
    # ========================================================
    mk_df.to_excel(
        writer,
        sheet_name="04_Mann_Kendall_Test",
        index=False
    )

    # ========================================================
    # Sheet 5
    # ========================================================
    trend_df.to_excel(
        writer,
        sheet_name="05_Trend_Summary",
        index=False
    )

print("="*70)
print("Arctic Sea Ice & Albedo Trend Analysis Workbook Created Successfully")
print("="*70)
print(f"\nSaved at:\n{OUTPUT_FILE}")
```
